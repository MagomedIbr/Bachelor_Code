```python
import os
import numpy as np
import matplotlib.pyplot as plt

import csv


labelss = []

resultss = []

errorr = []
with open('modeResultsUserUnfiltered.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        labelss.append(str(row[0]))
        resultss.append(float(row[1]))
        errorr.append(float(row[2]))

plt.figure(figsize=(2,2)) 
temp = np.array([resultss])
mean = temp.mean()
fig, ax = plt.subplots()    
width = 0.5 # the width of the bars 
ind = np.arange(len(resultss))  # the x locations for the groups
ax.barh(ind, resultss, width,xerr=errorr, color="tab:blue")
ax.set_yticks(ind+width/2)
ax.set_yticklabels(labelss,fontsize = 8, minor=False)
for i, v in enumerate(resultss):
    ax.text(v+1 , i + 0.1, str(v), color='black', fontweight='bold',size=9)
plt.title('Sprachmodus \n Durchschnitt: '+str(round(mean,2)))
plt.xlabel('Genauigkeit')
plt.ylabel('Klassifikator')
plt.margins(0.15)
plt.savefig(os.path.join('modeResultsUserUnfiltered.png'), dpi=300, format='png', bbox_inches='tight') # use format='svg' or 'pdf' for vectorial pictures
plt.show()
```

```python
from PIL import Image
import os
import PIL
import glob
import matplotlib.pyplot as plt

# path
path = "d"
listing = os.listdir(path)

#print (listing)

# getting all path+filename in a list
npath=[]
im=[]

for infile in listing:
    if infile.endswith("png"):
        #print(infile)
        im.append(infile)
        npath.append(os.path.join(path, infile))
print(im)
print(npath)

images = [Image.open(x) for x in im]
widths, heights = zip(*(i.size for i in npath))

total_width = sum(widths)
max_height = max(heights)

new_im = Image.new('RGB', (total_width, max_height))

x_offset = 0
for im in images:
    new_im.paste(im, (x_offset,0))
    x_offset += im.size[0]

new_im.show()
#new_im.save('test.jpg'))
```
