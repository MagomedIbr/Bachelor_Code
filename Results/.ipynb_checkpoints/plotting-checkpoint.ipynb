```python
import os
import numpy as np
import matplotlib.pyplot as plt

import csv


labelss = []

resultss = []

errorr = []
with open('UserResultsUser7Unf_without29.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        labelss.append(str(row[0]))
        resultss.append(float(row[1]))
        #errorr.append(float(row[2]))

#plt.figure(figsize=(20,2)) 
temp = np.array([resultss])
mean = temp.mean()
fig, ax = plt.subplots()    
width = 0.5 # the width of the bars 
ind = np.arange(len(resultss))  # the x locations for the groups
ax.barh(ind, resultss, width, color="tab:blue")
ax.set_yticks(ind+width/2)
ax.set_yticklabels(labelss,fontsize = 8, minor=False)
for i, v in enumerate(resultss):
    ax.text(v+0.001 , i + 0.001, str(v), color='black', fontweight='bold',size=8)
plt.title('User7 / Accuracy for User Identity \n Mean: '+str(round(mean,2)))
plt.xlabel('Accuracy')
plt.ylabel('Session')
plt.margins(0.15)
plt.savefig(os.path.join('UserResultsUser7Unf_without29.png'), dpi=600, format='png', bbox_inches='tight') # use format='svg' or 'pdf' for vectorial pictures
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

import csv


labels = []

results = []

error = []
with open('hqFreBandResultsUser.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        labels.append(str(row[0]))
        results.append(float(row[1]))
        error.append(float(row[2]))

#plt.figure(figsize=(5,10))        
fig, ax = plt.subplots()    
width = 0.6 # the width of the bars 
ind = np.arange(len(results))  # the x locations for the groups
ax.barh(ind, results, width,xerr=error, color="tab:blue")
ax.set_yticks(ind+width/2)
ax.set_yticklabels(labels, fontsize = 7, minor=False)
for i, v in enumerate(results):
    ax.text(v , i+0.2 , str(v), color='black', fontweight='bold',size=6)
plt.title('Accuracy for Speechmode ')
plt.xlabel('Accuracy')
plt.ylabel('Classifier')     
plt.savefig(os.path.join('hqFreBandResultsUser.png'), dpi=900, format='png', bbox_inches='tight') # use format='svg' or 'pdf' for vectorial pictures
plt.show()
```

```python
from PIL import Image
import os
import PIL
import glob
import matplotlib.pyplot as plt

# path
path = "d"
listing = os.listdir(path)


# getting all path+filename in a list
npath=[]
im=[]

for infile in listing:
    if infile.endswith(".png"):
        im.append(infile)
        npath.append(os.path.join(path, infile))

#creates a new empty image, RGB mode, and size 400 by 400.
new_im = Image.new('RGB', (2500,3000))

for i in range(0,2500,800):
    for j in range(0,3000, 500):
        im=Image.open(npath.pop(0))
        im.thumbnail((1000,1000))
        #paste the image at location i,j:
        new_im.paste(im, (i,j))
    #new_im.save('/media/test.png')
```
